# --- Day 16: Aunt Sue ---
https://adventofcode.com/2015/day/16

```python
import re

def getMFCSAMOutput():
	with open("MFCSAMOutput.txt") as file:
		return file.read()

def getAuntSues():
	with open("auntSues.txt") as file:
		return file.read()
```

```python
# Get the output of the MFCSAM as a dictionary
MFCSAMOutput = dict([[x.split(": ")[0], int(x.split(": ")[1])] for x in getMFCSAMOutput().split("\n")])

# Loop through all Sues
for sue in getAuntSues().split("\n"):
	sueNumber = re.findall(r'Sue (\d+):', sue)[0]
	
	# Loop through each compound that the MFCSAM tracks
	isCorrectSue = True
	for compound in MFCSAMOutput.keys():
		# Check to see if we remembered that compound
		compoundAmount = re.findall(f'{compound}: (\\d+)', sue)
		if compoundAmount:
			# Check to see if the compound matches
			if int(compoundAmount[0]) != MFCSAMOutput[compound]:
				isCorrectSue = False
	
	# If we find the correct Sue, print that and stop looking
	if isCorrectSue:
		print(f"The Aunt Sue that got the gift was number {sueNumber}")
		break
```

# --- Part Two ---

```python
# Get the output of the MFCSAM as a dictionary
MFCSAMOutput = dict([[x.split(": ")[0], int(x.split(": ")[1])] for x in getMFCSAMOutput().split("\n")])
greaterThan = ["cats", "trees"]
lessThan = ["pomeranians", "goldfish"]

# Loop through all Sues
for sue in getAuntSues().split("\n"):
	sueNumber = re.findall(r'Sue (\d+):', sue)[0]
	
	# Loop through each compound that the MFCSAM tracks
	isCorrectSue = True
	for compound in MFCSAMOutput.keys():
		# Check to see if we remembered that compound
		compoundAmount = re.findall(f'{compound}: (\\d+)', sue)
		# If we found a compound amount then try to disprove that this is the right Aunt Sue
		if compoundAmount:
			if compound in lessThan:
				# Check to see if the compound is greater than the output
				if int(compoundAmount[0]) >= MFCSAMOutput[compound]: isCorrectSue = False
			elif compound in greaterThan:
				# Check to see if the compound is less than the output
				if int(compoundAmount[0]) <= MFCSAMOutput[compound]: isCorrectSue = False
			else:
				# Check to see if the compound matches
				if int(compoundAmount[0]) != MFCSAMOutput[compound]: isCorrectSue = False
	
	# If we cannot disprove that this is the real Aunt Sue, then it must be the real one
	if isCorrectSue:
		print(f"The real Aunt Sue that got the gift was number {sueNumber}")
		break
```
