# --- Day 16: Aunt Sue ---
https://adventofcode.com/2015/day/16

```python
def getMFCSAMOutput():
	with open("MFCSAMOutput.txt") as file:
		return file.read()

def getAuntSues():
	with open("auntSues.txt") as file:
		return file.read()
```

```python
import re
```

```python
# Get the output of the MFCSAM as a dictionary
MFCSAMOutput = dict([[x.split(": ")[0], int(x.split(": ")[1])] for x in getMFCSAMOutput().split("\n")])

# Loop through all Sues
for sue in getAuntSues().split("\n"):
	sueNumber = re.findall(r'Sue (\d+):', sue)[0]
	
	# Loop through each compound that the MFCSAM tracks
	isCorrectSue = True
	for compound in MFCSAMOutput.keys():
		# Check to see if we remembered that compound
		compoundAmount = re.findall(f'{compound}: (\\d+)', sue)
		if compoundAmount:
			# Check to see if the compound matches
			if int(compoundAmount[0]) != MFCSAMOutput[compound]:
				isCorrectSue = False
	
	# If we find the correct Sue, print that and stop looking
	if isCorrectSue:
		print(f"The Aunt Sue that got the gift was number {sueNumber}")
		break
```
