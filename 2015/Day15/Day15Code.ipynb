```python
def getIngredients():
	with open("ingredients.txt") as file:
		return file.read()
```

```python
import re

ingredients = {}
pattern = r'(.*): (capacity -?\d), (durability -?\d), (flavor -?\d), (texture -?\d), (calories -?\d)'
for i in getIngredients().split("\n"):
	newIngredient = re.findall(pattern, i)[0]
	ingredients[newIngredient[0]] = dict([x.split() for x in newIngredient[1:]])

for key, value in ingredients.items(): print(f"{key}: {value}")
```
