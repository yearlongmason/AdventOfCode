```python
def getIngredients():
	with open("ingredients.txt") as file:
		return file.read()
```

```python
import re

class Ingredient:
	allIngredients = []
	
	def __init__(self, name: str, capacity: int, durability: int, flavor: int, texture: int, calories: int):
		self.name = name
		self.capacity = capacity
		self.durability = durability
		self.flavor = flavor
		self.texture = texture
		self.calories = calories
		Ingredient.allIngredients.append(self)

	def __str__(self):
		return f"{self.name} {self.capacity} {self.durability} {self.flavor} {self.texture} {self.calories}"
	


pattern = r'(.*): capacity (-?\d), durability (-?\d), flavor (-?\d), texture (-?\d), calories (-?\d)'
for i in getIngredients().split("\n"):
	Ingredient(*[val if i == 0 else int(val) for i, val in enumerate(re.findall(pattern, i)[0])])

for x in Ingredient.allIngredients: print(x)
```
