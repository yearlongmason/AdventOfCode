```python
def getLeastPresents():
	with open("leastPresents.txt") as file:
		return file.read()

leastPresents = int(getLeastPresents())

def getAllMultiples(num: int) -> list[int]:
	"""Get all multiples of a number"""
	allMultiples = [num]
	for i in range(1, num//2 + 1):
		if num % i == 0:
			allMultiples.append(i)
			
	return allMultiples

def getNumPresents(houseNumber: int) -> int:
	"""Get the number of presents delivered to the given house"""
	return sum(getAllMultiples(houseNumber)) * 10

# Set lowest and highest house using an exponential search
lowestHouse = 1
highestHouse = 1
while getNumPresents(highestHouse) < leastPresents: 
	highestHouse *= 2

# Binary search to find the lowest house number that has at least {leastPresents} presents
while abs(highestHouse - lowestHouse) != 1:
	midPoint = (highestHouse + lowestHouse) // 2

	if getNumPresents(midPoint) < leastPresents:
		lowestHouse = midPoint
	elif getNumPresents(midPoint) > leastPresents:
		highestHouse = midPoint

print(f"The first house to have at least {leastPresents} presents is house number {highestHouse}")
```
