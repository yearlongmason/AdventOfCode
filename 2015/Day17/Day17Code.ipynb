# --- Day 17: No Such Thing as Too Much ---
https://adventofcode.com/2015/day/17

```python
def getContainers():
	with open("containers.txt") as file:
		return file.read()
```

```python
containers = [int(x) for x in getContainers().split("\n")]

def findNumberOfCombinations(target: int, containersLeft: list[int], currentSum: int = 0) -> int:
	totalCombos = 0
	
	# Base case: we've reached the target sum
	if currentSum == target:
		return 1
	
	for container in containers:
		newContainersLeft = containersLeft
		newContainersLeft.remove(container)
		totalCombos += 1#findNumberOfCombinations()

print(findNumberOfCombinations(target=150, containersLeft=containers))
```
