```python
def getJson():
    with open("json.txt") as file: 
        return file.read()
```

```python
import re

print(f"Sum of all numbers in Santa's JSON: {sum([int(x) for x in re.findall("-?[0-9]+", getJson())])}")
```

# --- Part Two ---

```python
import json

santasJson = json.loads(getJson())

def explore(jsonObject: dict|list, isDict: bool = False) -> int:
    currentSum = 0
    newValues = jsonObject

    # Search through dictionary unless red exists
    if isDict:
        newValues = newValues.values()
        # If red is found then just return the current sum
        if "red" in newValues:
            return 0

    # Explore all of the new values
    for i in newValues:
        if isinstance(i, dict):
            currentSum += explore(i, isDict=True)
        elif isinstance(i, list):
            currentSum += explore(i)
        elif isinstance(i, int):
            currentSum += i

    return currentSum

print(f"Sum of all numbers in Santa's JSON (excluding red): {explore(santasJson, isDict=isinstance(santasJson, dict))}")
```
