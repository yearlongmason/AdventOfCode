# --- Day 14: Reindeer Olympics ---
https://adventofcode.com/2015/day/14

```python
def getReindeer():
    with open("reindeer.txt") as file:
        return file.read()
```

```python
reindeer = [{"Name": x.split()[0],
             "Speed": int(x.split()[3]),
             "Endurance": int(x.split()[6]),
             "Rest Time": int(x.split()[-2])} 
             for x in getReindeer().split("\n")]
reindeer
```
