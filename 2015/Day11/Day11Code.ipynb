# --- Day 11: Corporate Policy ---
https://adventofcode.com/2015/day/11

```python
def getPassword():
    with open("password.txt") as file:
        return file.read()
```

```python
import re

password = [ord(x) for x in [*getPassword()]] # Store password as list of ascci values

def incrementPassword(currentPassword: list[int]) -> list[int]:
    excludedLetters = [ord("i"), ord("l"), ord("o")]
    currentPassword[-1] += 1

    # If overflow
    for i in range(len(currentPassword) - 1,0,-1):
        if currentPassword[i] > ord("z"):
            currentPassword[i] = ord("a")
            if i: currentPassword[i - 1] += 1
    
    # Skips excluded letters
    currentPassword = [x + 1 if x in excludedLetters else x for x in currentPassword]

    return currentPassword

def containsStraight(password):
    for i in range(len(password) - 2):
        if password[i] == password[i + 1] - 1 and password[i] == password[i + 2] - 2: 
            return True
    return False

def getNumberOfPairs(password):
    pattern = f"({"|".join([x*2 for x in "abcdefghijklmnopqrstuvwxyz"])})"
    return len(re.findall(pattern, "".join([chr(x) for x in password])))

def isValidPassword(password: list[int]) -> bool:    
    return containsStraight(password) and getNumberOfPairs(password) >= 2

while not isValidPassword(password): 
    password = incrementPassword(password)

print(f"Next valid password: {"".join([chr(x) for x in password])}")
```

```python
password = [ord(x) for x in [*getPassword()]] # Store password as list of ascci values

numValidPasswords = 0
while numValidPasswords < 2:
    password = incrementPassword(password)
    if isValidPassword(password):
        numValidPasswords += 1

print(f"Next valid password: {"".join([chr(x) for x in password])}")
```
