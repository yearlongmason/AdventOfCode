# --- Day 13: Knights of the Dinner Table ---
https://adventofcode.com/2015/day/13

```python
def getHappinessMapping():
    with open("happinessMapping.txt") as file:
        return file.read()
```

```python
from collections import defaultdict
from itertools import permutations

# Cleaning happiness map
happinessMapping = getHappinessMapping().split("\n")
# Replace lose and gain with negative or empty for simplicity and remove "." at the end
happinessMapping = [x.replace("lose ", "-").replace("gain ", "")[0:-1] for x in happinessMapping]
happinessMapping = [[x.split()[0], int(x.split()[2]), x.split()[-1]] for x in happinessMapping]

# Create dictionary of dictionaries that act as a mapping from each person to their happiness mapping for each other person
happinessMap = defaultdict(lambda: defaultdict(int))
for person, effect, person2 in happinessMapping:
    happinessMap[person][person2] = effect
allPeople = list(happinessMap.keys()) # Get all individual people

def getHappinessScore(permutation: tuple) -> int:
    """Loop through from the person at the end to the second to last person
    return the sum of the mapping of every person to the next person"""
    return sum([happinessMap[permutation[i]][permutation[i+1]] + 
                happinessMap[permutation[i+1]][permutation[i]]
                 for i in range(-1, len(permutation)-1)])

for i in permutations(allPeople):
    print(i)
print(getHappinessScore(('David', 'Alice', 'Bob', 'Carol')))
print(f"Max happiness score: {max([getHappinessScore(x) for x in permutations(allPeople)])}")
```
