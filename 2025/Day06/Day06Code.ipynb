# --- Day 6: Trash Compactor ---
https://adventofcode.com/2025/day/6

```python
import re

def getInput():
	with open("mathHomework.txt") as file:
		return file.read()
```

```python
problems = getInput().split("\n")
numRows = len(problems) - 1
nums=[]

for row in problems[:-1]:
	nums.append([int(x) for x in re.findall(r"(\d+)", row)])
operations = re.findall(r"([+*])", problems[-1])

solutions = []
for i, operation in enumerate(operations):
	if operation == "+":
		solutions.append(sum([nums[x][i] for x in range(numRows)]))
	elif operation == "*":
		answer = 1
		for num in [nums[x][i] for x in range(numRows)]: answer *= num
		solutions.append(answer)

print(f"Grand total: {sum(solutions)}")
```

# --- Part Two ---

```python
problems = getInput().split("\n")
nums = problems[:-1]
operations = re.findall(r"([+*])", problems[-1])
numRowLength = len(nums[0])

def getNumbersUntilBlankCol(currentCol: int) -> tuple[list[int], int]:
	problemNumbers = []
	# Returns the list of numbers for the problem and the index to start at
	while {nums[x][currentCol] for x in range(len(nums))} != {" "}:
		problemNumbers.append("".join([nums[row][currentCol] for row in range(len(nums))]))
		currentCol += 1
		
		# If we're at the end break to avoid IndexError 
		if currentCol == numRowLength: break
	
	return [int(x.strip()) for x in problemNumbers][::-1], currentCol + 1

solutions = []
currentCol = 0
for operation in operations:
	problemNums, currentCol = getNumbersUntilBlankCol(currentCol)
	if operation == "+":
		solutions.append(sum(problemNums))
	elif operation == "*":
		answer = 1
		for num in problemNums: answer *= num
		solutions.append(answer)

print(f"Grand total: {sum(solutions)}")
```
