# --- Day 6: Trash Compactor ---
https://adventofcode.com/2025/day/6

```python
import re

def getInput():
	with open("mathHomework.txt") as file:
		return file.read()
```

```python
problems = getInput().split("\n")
numRows = len(problems) - 1

# Get operations and numbers for each problem
operations = re.findall(r"([+*])", problems[-1])
rows = []
for row in problems[:-1]:
	rows.append([int(x) for x in re.findall(r"(\d+)", row)])

# Get the numbers for each individual problems
problemNumbers = []
numProblems = len(rows[0])
for i in range(numProblems):
	problemNumbers.append([rows[j][i] for j in range(numRows)])

def solveProblem(problemNumbers: list[int], operation: str) -> int:
	# Get the answer for the problem given the numbers and the operation
	if operation == "*":
		answer = 1
		for num in problemNumbers: answer *= num
		return answer
	# If it's not "*" then it's "+"
	return sum(problemNumbers)

# Loop through each problem and operation and add the answer to the total
total = 0
for currentProblemNumbers, operation in zip(problemNumbers, operations):
	total += solveProblem(currentProblemNumbers, operation)

print(f"Grand total: {total}")
```

# --- Part Two ---

```python
problems = getInput().split("\n")
operations = re.findall(r"([+*])", problems[-1])
rows = problems[:-1]
rowLength = len(rows[0])

def getNumbersUntilBlankCol(currentCol: int) -> tuple[list[int], int]:
	# Returns the list of numbers for the problem and the index to start at
	problemNumbers = []
	# Keep looping until we find a column that is completely empty (a column that is all " ")
	while {rows[x][currentCol] for x in range(len(rows))} != {" "}:
		# Get the column as a string and move to the next column
		problemNumbers.append("".join([rows[row][currentCol] for row in range(len(rows))]))
		currentCol += 1
		
		# If we're at the end break to avoid IndexError 
		if currentCol == rowLength: break
	
	# Remove whitespace and cast as int for each problemNumber (and reverse)
	return [int(x.strip()) for x in problemNumbers][::-1], currentCol + 1

def solveProblem(problemNumbers: list[int], operation: str) -> int:
	# Get the answer for the problem given the numbers and the operation
	if operation == "*":
		answer = 1
		for num in problemNumbers: answer *= num
		return answer
	# If it's not "*" then it's "+"
	return sum(problemNumbers)

# Loop through each operation, get the numbers and solve the problem
total = 0
currentCol = 0
for operation in operations:
	# Get the next set of numbers, and change the column we start searching at next time
	problemNumbers, currentCol = getNumbersUntilBlankCol(currentCol)
	total += solveProblem(problemNumbers, operation)

print(f"Grand total: {total}")
```
