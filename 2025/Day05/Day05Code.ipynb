# --- Day 5: Cafeteria ---
https://adventofcode.com/2025/day/5

```python
from copy import deepcopy

def getInput():
	with open("IDs.txt") as file:
		return file.read()
```

```python
idRanges, availableIds = getInput().split("\n\n")
idRanges = [[int(y) for y in x.split("-")] for x in idRanges.split("\n")]
availableIds = [int(x) for x in availableIds.split("\n")]

# For each ID check if it's in between any ranges
freshIngredients = 0
for id in availableIds:
	for lowerBound, upperBound in idRanges:
		if id >= lowerBound and id <= upperBound:
			freshIngredients += 1
			break
		
print(f"Number of available fresh ingredient IDs: {freshIngredients}")
```

# --- Part Two ---

```python
idRanges = getInput().split("\n\n")[0]
idRanges = [[int(y) for y in x.split("-")] for x in idRanges.split("\n")]

def combineRanges(currentRanges: list[list[int, int]]) -> tuple[list[list[int, int]], bool]:
	# Returns new ranges and whether or not a change was made
	for i, firstRange in enumerate(currentRanges):
		for j, secondRange in enumerate(currentRanges):
			if i == j: continue # Skip if on the same range
			firstLower, firstUpper = firstRange
			secondLower, secondUpper = secondRange

			# Check for same range at different spots
			if firstRange == secondRange:
				# Just remove one range (doesn't matter which)
				return [x for index, x in enumerate(currentRanges) if index != i], True

			# Look for complete overlap
			if firstLower <= secondLower and firstUpper >= secondUpper:
				# Return all except second range
				return [x for index, x in enumerate(currentRanges) if index != j], True

			# firstLower < secondLower < firstUpper < secondUpper = overlap
			if firstLower <= secondLower and secondLower <= firstUpper and firstUpper <= secondUpper:
				# New range will be firstLower -> secondUpper
				currentRanges.append([firstLower, secondUpper])
				return [x for index, x in enumerate(currentRanges) if index not in [i,j]], True

	# We made it through the whole thing so we're done combining
	return currentRanges, False

# Keep combining until no more changes are being made		
updateMade = True
while updateMade:
	idRanges, updateMade = combineRanges(deepcopy(idRanges))

# Get total number of IDs
totalFreshIds = 0
for lowerBound, upperBound in idRanges:
	totalFreshIds += upperBound - lowerBound + 1
	
print(f"Total number of fresh IDs: {totalFreshIds}")
```
