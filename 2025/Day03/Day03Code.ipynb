# --- Day 3: Lobby ---
https://adventofcode.com/2025/day/3

```python
def getInput():
	with open("banks.txt") as file:
		return file.read()
```

```python
totalOutputJoltage = 0
for bank in getInput().split("\n"):
	# Get the highest digit in the bank excluding the last digit (this is the first digit)
	# Get the highest of all digits that come after the first digit
	firstDigit = max(bank[:-1])
	secondDigit = max(bank[bank.index(firstDigit)+1:])
	totalOutputJoltage += int(firstDigit + secondDigit)

print(f"Total output joltage: {totalOutputJoltage}")
```

# --- Part Two ---

```python
def getNextBestNumber(currentBank: str, digitsRemaining: int) -> tuple[str, str]:
	# If we're on our last digit just return the highest remaining digit
	if digitsRemaining == 1:
		return "", str(max([int(x) for x in currentBank]))
	
	# Get the highest digit from the available digits 
	# (all except the last n, where n is the number of digits remaining)
	nextDigit = max(currentBank[:(digitsRemaining-1)*-1])
	# Get the remaining bank (all digits after the newest digit)
	remainingBank = currentBank[currentBank.index(nextDigit)+1:]
	return remainingBank, nextDigit

totalOutputJoltage = 0
for bank in getInput().split("\n"):
	# Keep track of current joltage
	joltage = ""
	# Loop through the digits remaining
	for digitsRemaining in range(12,0,-1):
		# Get the next digit and the rest of the bank
		bank, nextDigit = getNextBestNumber(bank, digitsRemaining)
		joltage += nextDigit

	totalOutputJoltage += int(joltage)

print(f"Total output joltage: {totalOutputJoltage}")
```
