# --- Day 1: Secret Entrance ---
https://adventofcode.com/2025/day/1

```python
def getInput():
	with open("instructions.txt") as file:
		return file.read()
```

```python
# Turn the instructions into a list of the number of rotations
# Negative if the instruction is to turn left, otherwise positive
instructions = getInput()
instructions = [[x[0], int(x[1:])] for x in instructions.split("\n")]
instructions = [x[1] if x[0] == "R" else x[1]*-1 for x in instructions]

numZeros = 0
dial = 50
for rotations in instructions:
	# Add the rotation and mod by 100 to account for dial overflow or underflow
	dial += rotations
	dial %= 100
	if dial == 0:
		numZeros += 1
		
print(f"Number of times the dial lands on 0: {numZeros}")
```

# --- Part Two ---

```python
# Turn the instructions into a list of the number of rotations
# Negative if the instruction is to turn left, otherwise positive
instructions = getInput()
instructions = [[x[0], int(x[1:])] for x in instructions.split("\n")]

numZeros = 0
dial = 50
for direction, rotations in instructions:
	# Optimization that accounts for rotations over 100
	numZeros += rotations//100
	rotations %= 100

	# Loop through each rotation in the instruction
	for _ in range(rotations):
		# Move dial and account for dial overflow/underflow
		dial += 1 if direction == "R" else -1
		dial %= 100
		if dial == 0:
			numZeros += 1

print(f"Number of times the dial hits 0: {numZeros}")
```
