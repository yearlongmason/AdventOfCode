# --- Day 1: Secret Entrance ---
https://adventofcode.com/2025/day/1

```python
def getInput():
	with open("instructions.txt") as file:
		return file.read()
```

```python
instructions = getInput()

numZeros = 0
dial = 50
for i in instructions.split("\n"):
	if i[0] == "R":
		for i in range(int(i[1:])):
			dial += 1
			if dial == 100:
				dial = 0
	else:
		for i in range(int(i[1:])):
			dial -= 1
			if dial == -1:
				dial = 99
	
	if dial == 0:
		numZeros += 1
		
print(f"Number of times the dial lands on 0: {numZeros}")
```
