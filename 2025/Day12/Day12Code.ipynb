# --- Day 12: Christmas Tree Farm ---
https://adventofcode.com/2025/day/12

```python
import re

def getInput():
	with open("presents.txt") as file:
		return file.read()
```

```python
presentSizes = {i: present.count("#") for i, present in enumerate(getInput().split("\n\n")[:-1])}
trees = getInput().split("\n\n")[-1].splitlines()

def getNumCoveredSpots(presentSizes: dict[int, int], presentCounts: list[int]) -> int:
	"""Returns the total number of spaces that the presents will take up"""
	totalCoveredSpaces = 0
	for i, numPresents in enumerate(presentCounts):
		totalCoveredSpaces += presentSizes[i] * numPresents
	
	return totalCoveredSpaces

validRegions = 0
for tree in trees:
	# Get all inforamtion from a tree
	presentCounts = list(map(int, re.findall(r'\d+', tree.split(":")[1])))
	length, width = list(map(int, re.findall(r'(\d+)x(\d+)', tree)[0]))
	regionSize = length * width

	# Determine if the space can hold the total number of spots covered by the presents
	if regionSize >= getNumCoveredSpots(presentSizes, presentCounts):
		validRegions += 1

print(f"Number of trees that can fit the given presents under them: {validRegions}")
```
