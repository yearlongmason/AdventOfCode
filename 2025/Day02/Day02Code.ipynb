# --- Day 2: Gift Shop ---
https://adventofcode.com/2025/day/2

```python
import re

def getInput():
	with open("IDRanges.txt") as file:
		return file.read()
```

```python
ranges = re.findall(r'(\d+)-(\d+)', getInput())
ranges = [map(int, x) for x in ranges]

def isInvalid(id: str) -> bool:
	# Check to see if id is invalid
	return id[:len(id)//2] == id[len(id)//2:]

invalidIdSum = 0
# Loop through each range
for lower, upper in ranges:
	# Loop through each id
	for id in range(lower, upper+1):
		if isInvalid(str(id)):
			invalidIdSum += id

print(f"Sum of all invalid IDs: {invalidIdSum}")
```

# --- Part Two ---

```python
ranges = re.findall(r'(\d+)-(\d+)', getInput())
ranges = [map(int, x) for x in ranges]

def isInvalid(id: str) -> bool:
	# Return True if the id is invalid otherwise return False
	for chunkLength in range(1, (len(id)//2)+1):
		# Check to make sure it's actually divisible by i
		if len(id) % chunkLength != 0:
			continue
		# Split id into even sized chunks in a set comprehension to remove duplicates
		# If the length of the set is 1 then all the chunks were the same making the id invalid
		if len({id[chunkStart:chunkStart+chunkLength] for chunkStart in range(0,len(id),chunkLength)}) == 1:
			return True
	return False

invalidIdSum = 0
# Loop through each range
for lower, upper in ranges:
	# Loop through each id
	for id in range(lower, upper+1):
		if isInvalid(str(id)):
			invalidIdSum += id

print(f"Sum of all invalid IDs: {invalidIdSum}")
```

# Part 2 Using Regex
Bonus: Much faster!

```python
ranges = re.findall(r'(\d+)-(\d+)', getInput())
ranges = [map(int, x) for x in ranges]

def isInvalid(id: str) -> bool:
	return re.match(r'\b(\d+)\1{1,' + str(len(id)) + r'}\b', id)

invalidIdSum = 0
# Loop through each range
for lower, upper in ranges:
	# Loop through each id
	for id in range(lower, upper+1):
		if isInvalid(str(id)):
			invalidIdSum += id

print(f"Sum of all invalid IDs: {invalidIdSum}")
```
